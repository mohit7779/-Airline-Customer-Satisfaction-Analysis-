```python
# set style for plot grid
sns.set(style = "whitegrid")
plt.figure(figsize = (12,6))
```

# age distribution

```python

plt.figure(figsize = (12,6))
plt.title("Age Distribution")
sns.histplot(x = "Age" , data = df,bins = 30,kde = True , color = "skyblue")
```

# Satisfaction Count


```python
sns.countplot(x = "satisfaction",data = df , palette = "viridis")
plt.title("Satisfaction Count",fontsize = 20)
```

# Flight Distance Distribution by Satisfaction

```python
sns.histplot(x = "Flight Distance",hue = "satisfaction",data = df,bins = 30,kde = True)
plt.title("Flight Distance Distribution by Satisfaction")
```

```python
plt.title("Gender Distribution by Satisfaction")
sns.countplot(x = "Gender", hue = "satisfaction",data = df)
plt.figure(figsize = (12,6))
```

```python
sns.countplot(x = "Customer Type",hue = "Gender",data = df)
```

```python
avg_delays = df.groupby('satisfaction')[['Departure Delay in Minutes', 'Arrival Delay in Minutes']].mean().reset_index()

avg_delays.plot(x='satisfaction', kind='bar', ax=plt.gca())
plt.title("Average Delay Times by Satisfaction")
plt.ylabel("Average Delay (Minutes)")
plt.xticks(rotation=0)
```

```python
df["Class"].value_counts().plot.pie(autopct="%0.2f%%",label = "Class")
plt.title("Distribution of class By Percentage Category")
```

```python
sns.countplot(x = "Class",hue = "Type of Travel",data = df)
```

```python
sns.boxplot(x = "Customer Type", y = "Flight Distance",data = df,palette = "Set2")
plt.title("Flight Distance by Customer Type")
```
